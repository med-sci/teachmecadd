```python
import sys
import os
import pandas as pd
from collections import Counter
import warnings
sys.path.append("../")
warnings.filterwarnings("ignore")
from tcad.tools.datatools import DataLoader, preprocess
```

### Loading activity data for CHEMBL1871 target.

```python
data_loader = DataLoader('CHEMBL1871')
data_loader.load_data()
data = data_loader.get_data()

data.head()
```

### We interested only in subset of columns with useful information

```python
columns = [
    'assay_type',
    'canonical_smiles', 
    'molecule_chembl_id', 
    'standard_units',
    'standard_value', 
    'target_organism',
    'type',
    ]

data = preprocess(data, *columns)
```

```python
filename = "androgen_data.csv"
datapath=os.path.join('../data', filename)
data = data.query("assay_type=='B'&type=='IC50'")

data.to_csv(datapath, index=False)
data.head()
```
